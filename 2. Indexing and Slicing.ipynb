```python
df2 = pd.DataFrame({'a': ['one', 'one', 'two', 'three', 'two', 'one', 'six'],
   .....:                     'b': ['x', 'y', 'y', 'x', 'y', 'x', 'x'],
   .....:                     'c': np.random.randn(7)})

df2
```

Filtering with map can also be done:

```python
df2['a'].map(lambda x: x.startswith('t'))
```

Similar to, but faster:

```python
#str has to be specified so string methods can be applied to a series
df2['a'].str.startswith('t')
```

Expanding on this we can use the boolean map to select data

```python
#previously demonstrated method
start_time = time.time()
df2[df2.a.str.startswith('t')]
print("--- %s seconds ---" % (time.time() - start_time))
```

Alternative

```python
start_time = time.time()
criteria = df2['a'].map(lambda x: x.startswith('t'))
df2[criteria]
print("--- %s seconds ---" % (time.time() - start_time))
```

2nd Map method

```python
start_time = time.time()
df2[df2['a'].map(lambda x: x.startswith('t'))]
print("--- %s seconds ---" % (time.time() - start_time))
```

More complex criterias

```python
df2[criteria & (df2['b'] == 'x')]
```

```python
df2.loc[criteria & (df2['b'] == 'x'), lambda x: x.columns == 'a']
```

```python
df2.loc[criteria & (df2['b'] == 'x'), ['a', 'b']]
```
